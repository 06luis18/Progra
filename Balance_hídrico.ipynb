```
def b(data):
  lista = []
  for i in range(len(data)):
    if data.Dias[i] == 28:
      lista.append(round((58.93 * (4 * data.Dias[i] + 1) * data.T_med[i])/(365 * 4), 1))
    else:
      lista.append(round((58.93 * data.Dias[i] * data.T_med[i])/365, 1))
  data["ETP"] = lista
  data["Balance_Hidrico"] = data.Pp - data.ETP
  lista2 = []
  for i in range(48):
    if data.ETP[i] > data.Pp[i]:
      if data.Balance_Hidrico[i -1] > 0:
        lista2.append(round(data.Pp[i] + data.Balance_Hidrico[i - 1], 1))
      else:
        lista2.append(round(data.Pp[i], 1))
    else:
      lista2.append(round(data.ETP[i], 1))
  lista2
  data["ETR"] = lista2
  return(data)
```
